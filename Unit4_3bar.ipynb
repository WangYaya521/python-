```python
import matplotlib.pyplot as plt
import numpy as np

# 数据准备
platforms = ["豆瓣", "猫眼", "淘票票", "时光网", "IMDB", "烂番茄", "知乎电影", "豆瓣万人评"]
xiaoshenke = [9.7, 9.8, 9.7, 9.5, 9.3, 9.1, 9.2, 9.6]
bawangbieji = [9.6, 9.6, 9.5, 9.3, 8.9, 8.8, 9.0, 9.5]

y = np.arange(len(platforms))  # y轴位置（对应每个平台）
bar_height = 0.35  # 水平柱子的"高度"（实际是垂直方向的厚度）

# 绘制水平分组柱状图
plt.barh(y - bar_height/2, xiaoshenke, height=bar_height, color="blue", label="肖申克的救赎")
plt.barh(y + bar_height/2, bawangbieji, height=bar_height, color="orange", label="霸王别姬")

# 设置标签和标题
plt.yticks(y, platforms)  # y轴显示平台名称
plt.xlabel("评分")
plt.ylabel("评分平台")
plt.title("电影评分水平对比柱状图")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 散点图 scatter(x , y , s , c) x表示x轴位置，y表示y轴位置，s表示散点大小，c表示散点颜色

fig = plt.figure(figsize = (10,4))

plt.scatter(movies_score["电影名称"] , movies_score["猫眼"] , s = 10 , c = "red" , label = "猫眼评分散点图")

plt.xticks(rotation = 90 , fontsize = 5)
plt.xlabel("movie")
plt.yticks(np.arange(7 , 10.5 , 0.5))
plt.ylabel("score")

plt.title("猫眼评分的散点图")

plt.legend(loc = "upper right" , frameon = True , fontsize = 12)

plt.show()
```

```python
# 频率直方图
movies_cateye = movies_score["猫眼"]

# 2. 直接调用 .hist() 画直方图
fig = plt.figure(figsize = (6,4))

plt.hist(movies_cateye , bins = 30 , color = "purple" , edgecolor = "black" , range=(7,10) , label="猫眼评分频率分布")

plt.xlabel("评分区间")
plt.xticks(np.arange(7 , 10 , 0.1) , rotation = 60)
plt.ylabel("频数")
plt.legend(loc = "upper left" , frameon = True , fontsize = 12)

# 3. 显示图表
plt.show()
```

```python

# 2. 创建画布和子图
fig = plt.figure(figsize=(16, 6))  # 加宽画布，避免标签重叠
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 3. 绘制 ax1（豆瓣评分，前10部）
ax1.bar(
    movies_score["电影名称"][0:10],
    movies_score["豆瓣"][0:10],
    width = 0.4,
    color="blue",
    label="豆瓣评分"
)
ax1.set_xticklabels(movies_score["电影名称"][0:10], rotation=70, fontsize=10)
ax1.set_yticks(np.arange(0,11,1))
ax1.set_xlabel("电影")
ax1.set_ylabel("评分")
ax1.set_title("豆瓣评分的柱状图")
ax1.legend()

# 4. 绘制 ax2（猫眼评分，前20部，建议也改成前10部保持一致）
ax2.bar(
    movies_score["电影名称"][0:10],  # 改成和左边一样的数量，避免拥挤
    movies_score["猫眼"][0:10],
    color="red",
    label="猫眼评分"
)
ax2.set_xticklabels(movies_score["电影名称"][0:10], rotation=70, fontsize=10)
ax2.set_yticks(np.arange(0,11,1))
ax2.set_xlabel("电影")
ax2.set_ylabel("评分")
ax2.set_title("猫眼评分的柱状图")
ax2.legend()

# 5. 调整布局，防止标签被截断
plt.tight_layout()
plt.show()
```

```python
# 盒图.boxplot(data , vert , patch_artist) data表示数据，vert表示是否垂直显示，patch_artist表示是否填充颜色
fig = plt.figure(figsize = (6,4))

plt.boxplot(movies_score["猫眼"])

plt.xticks([1] , ["猫眼评分"]) # 设置x轴刻度标签

plt.yticks(np.arange(7 , 10.5,0.5))

plt.show()
```
