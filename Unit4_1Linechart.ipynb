```python
import pandas as pd
import numpy as np
from pandas import Series
from tabulate import tabulate
from pypinyin import lazy_pinyin

unemployees_rate = pd.read_csv(r"D:\JupyterNotebook\PythonFile\matplotlib\American_unemployees_rating.csv")
unemployees_rate["date"] = pd.to_datetime(unemployees_rate["date"])
print(unemployees_rate.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.plot()


plt.show()
```

```python
sns.set_style("whitegrid")

first_twelve = unemployees_rate[0:300]

# plt.plot(横坐标，纵坐标)

plt.plot(first_twelve["date"] , first_twelve["unemployment_rate"])

# 横坐标显示完全.xticks(rotation = 30)标签以 30° 角顺时针旋转
# 纵坐标.yticks(rotation = ...)

plt.xticks(rotation = 30)

sns.despine(left = True , bottom = True)

sns.set_context("notebook" , font_scale = 3 , rc = {"lines.linewidth" : 3})

plt.show()
```

```python
first_twelve = unemployees_rate[0:500]

# plt.plot(横坐标，纵坐标)

plt.plot(first_twelve["date"] , first_twelve["unemployment_rate"])
plt.xticks(rotation = 30)

# x轴，y轴分别表示什么.xlabel  .ylabel

plt.xlabel("Month")
plt.ylabel("unemployees_rating")

# 图头名.title

plt.title("American 1988-2025's unemployees_rate")
plt.show()
```
