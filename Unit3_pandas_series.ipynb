```python
import pandas as pd
import numpy as np
from pandas import Series
from tabulate import tabulate

movie_ratings = pd.read_csv(r"D:\JupyterNotebook\PythonFile\Pandas\movie_ratings.csv" , encoding = "gbk")

film_name_series = movie_ratings["电影名称"]
film_name_series[23:34]
print(type(film_name_series))

the_tenth_row = movie_ratings.loc[0]
print(the_tenth_row)

the_one_five_row = movie_ratings[1:5]
print(type(the_one_five_row))
```

```python
# series.values可以将pandas中series转化为向量
film_names = film_name_series.values
# print(type(film_names))
# print(film_names)

scores_series = movie_ratings["淘票票"]

scores = scores_series.values

series_custom = Series(scores , index = film_names)
# series_custom
# series_custom[2:20:3]
series_custom[["阿甘正传" , "肖申克的救赎"]]
```
