```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pypinyin import lazy_pinyin

unemployees_rate = pd.read_csv(r"D:\JupyterNotebook\PythonFile\matplotlib\American_unemployees_rating.csv")
unemployees_rate["date"] = pd.to_datetime(unemployees_rate["date"])

rate_1988_1998 = unemployees_rate.head(120)
rate_1998_2008 = unemployees_rate[121:240]
rate_2008_2018 = unemployees_rate[241:360]
rate_2018_2024 = unemployees_rate[360:]

fig = plt.figure(figsize = (8 , 10))

ax1 = fig.add_subplot(2,2,1)
plt.plot(rate_1988_1998["date"] , rate_1988_1998["unemployment_rate"] , color = "blue" )
plt.xlabel("year")
plt.xticks(rotation = 30)
plt.ylabel("unemployment_rate")
plt.title("1988-1998's unemployees_rate")

ax2 = fig.add_subplot(2,2,4)
plt.plot(rate_1998_2008["date"] , rate_1998_2008["unemployment_rate"] , color = "red" )
plt.xlabel("year")
plt.xticks(rotation = 30)
plt.ylabel("unemployment_rate")
plt.title("1998-2008's unemployees_rate")

plt.show()
```

```python
# 一张图中两条线 

fig = plt.figure(figsize = (8,6))

plt.plot(rate_2008_2018["date"] , rate_2008_2018["unemployment_rate"] , color = "blue")
plt.plot(rate_2018_2024["date"] , rate_2018_2024["unemployment_rate"] , c = "red")

plt.show()
```

```python
import numpy as np
import pandas as pd
from pypinyin import lazy_pinyin
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'WenQuanYi Micro Hei']  # 优先用黑体/微软雅黑/文泉驿
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题


foods_info = pd.read_csv(r"D:\JupyterNotebook\PythonFile\Pandas\foods_info.csv" , encoding = "gbk")

# 水分和总脂肪对比

food_water = foods_info["水分_(g)"]
food_fat = foods_info["总脂肪_(g)"]
food_danbaizhi = foods_info["蛋白质_(g)"]
food_huifen = foods_info["灰分_(g)"]
food_zongtang = foods_info["总糖_(g)"]


fig = plt.figure(figsize = (8,6))


# 绘制各成分折线图

plt.plot(foods_info["食物名称"][0:20] , food_water[0:20] ,  label = "water")

plt.plot(foods_info["食物名称"][0:20] , food_fat[0:20] ,label = "fat")

plt.plot(foods_info["食物名称"][0:20] , food_danbaizhi[0:20] , label = "danbaizhi")

plt.plot(foods_info["食物名称"][0:20] , food_huifen[0:20] , label = "huifen")

plt.plot(foods_info["食物名称"][0:20] , food_zongtang[0:20] , label = "zongtang")

# 显示图例标注

plt.legend(loc = "upper right" , frameon = True , fontsize = 10) 

# 调整横纵坐标，纵坐标刻度细分

plt.xticks(rotation = 60)
plt.yticks(np.arange(0,100,5) , rotation = 30)


plt.xlabel("foods")
plt.ylabel("含量(g)")
plt.title("食物中各成分含量占比")





plt.show()
```
