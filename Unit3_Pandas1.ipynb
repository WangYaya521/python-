```python
import pandas as pd
import numpy as np

employees_info = pd.read_csv(r"D:\JupyterNotebook\PythonFile\Pandas\employees.csv" , encoding = "gbk")

employees_info
```

```python
from tabulate import tabulate
print(tabulate(employees_info.head(3), headers='keys', tablefmt="psql"))

print("-----------------------------")

employees_info.tail(4)
```

```python
age = employees_info["年龄"]
# print(age[0:20])

age_is_null = pd.isnull(age) # 判断age中哪些值是空值，返回一个布尔类型的Series,True表示对应位置的值是空值，False表示对应位置的值不是空值
print(age_is_null)

age_null_true= age[age_is_null]
print(age_null_true)

age_null_count = len(age_null_true)
print(age_null_count)

age_sum = age.sum()
print(age_sum)

age_mean = age_sum / len(age) # 计算平均值，age_sum是年龄的总和，len(age)是年龄的数量,但是不会忽略空值，导致平均数降低

print(age_mean)

age_mean1 = age.mean() # 直接使用mean()方法计算平均值，mean()方法会自动忽略空值，并计算非空值的平均值
print(age_mean1)

```

```python
dengji = [1,2,3]
dnegji_of_employees = employees_info["等级"]
money_for_dengji = {}
people_for_dengji = {}
for jibie in dengji:
    dengji_employees_rows = employees_info[dnegji_of_employees == jibie]
    dengji_for_money = dengji_employees_rows["月薪"]
    dengji_money_mean = dengji_for_money.mean()
    money_for_dengji[jibie] = dengji_money_mean
print(money_for_dengji)

print("===========================")

for jibie in dengji:
    dengji_employees_rows = employees_info[dnegji_of_employees ==jibie]
    dengji_of_people = dengji_employees_rows["姓名"]
    people_for_dengji[jibie] = dengji_of_people
print(people_for_dengji)
```
